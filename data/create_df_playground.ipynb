```python
import pandas as pd
```

```python
ghi_df = pd.read_csv('./GHI_data.csv')
ghi_df['datetime'] = ghi_df['datetime'].str.replace(r'[^\x20-\x7E]', '', regex=True)
ghi_df['datetime'] = pd.to_datetime(ghi_df['datetime'], format='%d.%m.%Y %H:%M:%S.%f')
ghi_df = ghi_df.drop(columns=['time'])
ghi_df.head()
```

```python
# create the weather table with all info 
df = pd.read_csv('./UCY_SM_historical.csv')
df = df.drop_duplicates()
df['datetime'] = df['datetime'].str.replace(r'[^\x20-\x7E]', '', regex=True)
df['datetime'] = pd.to_datetime(df['datetime'], format='%d.%m.%Y %H:%M:%S.%f')
weather_df_0 = pd.read_csv('./UCY_SM_Weather_0.csv')
weather_df_1 = pd.read_csv('./UCY_SM_Weather_1.csv')
weather_df_2 = pd.read_csv('./UCY_SM_Weather_2.csv')
weather_df_3 = pd.read_csv('./UCY_SM_Weather_3.csv')
weather_df_4 = pd.read_csv('./UCY_SM_Weather_4.csv')
weather_df = pd.concat([weather_df_0, weather_df_1, weather_df_2, weather_df_3, weather_df_4], ignore_index=True)
weather_df['datetime'] = pd.to_datetime(weather_df['datetime'])
print(df.shape)
print(weather_df.shape)
df_all = pd.merge(df, weather_df, on='datetime', how='left')
df_all = df_all.sort_values('datetime')
print(df_all.shape)

df.head(10)
```

```python
# add ghi data
df_all = pd.merge(df_all, ghi_df, on='datetime', how='left')
print(df_all.shape)
df_all.head(10)
```

```python
df_all.sort_values('datetime')
df_all = df_all.drop_duplicates()
print(df_all.shape)
df_all.head(10)
df_all.to_csv('./UCY_SM_historical_with_weather.csv', index=False)
```
